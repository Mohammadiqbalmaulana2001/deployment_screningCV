```python
import requests

url = 'http://localhost:5000/predict_pdf'  # Ganti dengan URL server Flask Anda

files = {'file': open('business.pdf', 'rb')}
# Ganti dengan path sebenarnya menuju file PDF yang ingin Anda43 uji

response = requests.post(url, files=files)

if response.status_code == 200:
    result = response.json()
    predicted_category = result.get('predicted_category')
    print(f"Predicted Category from Uploaded File: {predicted_category}")
else:
    print("Error occurred:", response.text)
```
